```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('Arrival Events/Friday/HS19/strep_0.pickle', 'rb') as f:
    HS = pickle.load(f)
with open('Arrival Events/Friday/LS19/strep_0.pickle', 'rb') as f:
    LS = pickle.load(f)
events = HS + LS
events.sort(key=lambda x: x.time)
```

```python
plt.hist([e.time for e in events if e.time < 24*3600], bins = 24)
```

```python
data = np.array([[e.time for e in events if e.time < 24*3600], [e.severity for e in events if e.time < 24*3600], [e.disposition_code for e in events if e.time < 24*3600]]).T
df = pd.DataFrame(data, columns = ['time', 'severity', 'disposition'])
```
